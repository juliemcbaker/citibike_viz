```python
# import dependencies
import pandas as pd
import os
```

## Function to convert column headers to current format

```python
def fix_columns(my_frame):
    my_frame.rename(columns= {'Trip Duration': 'tripduration', 'Start Time': 'starttime', 'Stop Time': 'stoptime',
        'Start Station ID': 'start station id', 'Start Station Name': 'start station name', 
        'Start Station Latitude': 'start station latitude', 'Start Station Longitude': 'start station longitude', 
        'End Station ID': 'end station id', 'End Station Name': 'end station name', 'End Station Latitude': 'end station latitude', 
        'End Station Longitude': 'end station longitude', 'Bike ID': 'bikeid', 'User Type': 'usertype', 'Birth Year': 'birth year', 
        'Gender': 'gender'}, inplace=True) 
    print(my_frame.columns)
```

## 2016 Data
* Capital Titles

#### convert csv files to dataframes

```python
Jan2016_df = pd.read_csv('data/2016/JC-20161-citibike-tripdata.csv')
Feb2016_df = pd.read_csv('data/2016/JC-20162-citibike-tripdata.csv')
Mar2016_df = pd.read_csv('data/2016/JC-20163-citibike-tripdata.csv')
Apr2016_df = pd.read_csv('data/2016/JC-201604-citibike-tripdata.csv')
May2016_df = pd.read_csv('data/2016/JC-201605-citibike-tripdata.csv')
Jun2016_df = pd.read_csv('data/2016/JC-201606-citibike-tripdata.csv')
Jul2016_df = pd.read_csv('data/2016/JC-201607-citibike-tripdata.csv')
Aug2016_df = pd.read_csv('data/2016/JC-201608-citibike-tripdata.csv')
Sep2016_df = pd.read_csv('data/2016/JC-201609-citibike-tripdata.csv')
Oct2016_df = pd.read_csv('data/2016/JC-201610-citibike-tripdata.csv')
Nov2016_df = pd.read_csv('data/2016/JC-201611-citibike-tripdata.csv')
Dec2016_df = pd.read_csv('data/2016/JC-201612-citibike-tripdata.csv')
```

#### reformat columns

```python
list_2016 = [Jan2016_df, Feb2016_df, Mar2016_df, Apr2016_df, May2016_df, Jun2016_df, Jul2016_df, Aug2016_df, 
    Sep2016_df, Oct2016_df, Nov2016_df, Dec2016_df]

for frame in list_2016:
    fix_columns(frame)
```

### concatenate dfs
* into quarters
* full year

```python
Q1_2016 = pd.concat([Jan2016_df, Feb2016_df, Mar2016_df], axis=0)
Q2_2016 = pd.concat([Apr2016_df, May2016_df, Jun2016_df], axis=0)
Q3_2016 = pd.concat([Jul2016_df, Aug2016_df, Sep2016_df], axis=0)
Q4_2016 = pd.concat([Oct2016_df, Nov2016_df, Dec2016_df], axis=0)
all_2016 = pd.concat([Q1_2016, Q2_2016, Q3_2016, Q4_2016], axis=0)
all_2016
```

### load 2016 dataframes into csv

```python
Q1_2016.to_csv('data/for_tableau/q1_2016.csv')
Q2_2016.to_csv('data/for_tableau/q2_2016.csv')
Q3_2016.to_csv('data/for_tableau/q3_2016.csv')
Q4_2016.to_csv('data/for_tableau/q4_2016.csv')
all_2016.to_csv('data/for_tableau/all_2016.csv')
```
