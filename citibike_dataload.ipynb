## 2016 Data

#### convert csv files to dataframes

```python
Jan2016_df = pd.read_csv('data/2016/JC-20161-citibike-tripdata.csv')
Feb2016_df = pd.read_csv('data/2016/JC-20162-citibike-tripdata.csv')
Mar2016_df = pd.read_csv('data/2016/JC-20163-citibike-tripdata.csv')
Apr2016_df = pd.read_csv('data/2016/JC-201604-citibike-tripdata.csv')
May2016_df = pd.read_csv('data/2016/JC-201605-citibike-tripdata.csv')
Jun2016_df = pd.read_csv('data/2016/JC-201606-citibike-tripdata.csv')
Jul2016_df = pd.read_csv('data/2016/JC-201607-citibike-tripdata.csv')
Aug2016_df = pd.read_csv('data/2016/JC-201608-citibike-tripdata.csv')
Sep2016_df = pd.read_csv('data/2016/JC-201609-citibike-tripdata.csv')
Oct2016_df = pd.read_csv('data/2016/JC-201610-citibike-tripdata.csv')
Nov2016_df = pd.read_csv('data/2016/JC-201611-citibike-tripdata.csv')
Dec2016_df = pd.read_csv('data/2016/JC-201612-citibike-tripdata.csv')
```

#### check dataframes

### concatenate dfs

* Into quarters & then full year

```python
Q1_2016 = pd.concat([Jan2016_df, Feb2016_df, Mar2016_df], axis=0)
Q1_2016
```

```python
Q2_2016 = pd.concat([Apr2016_df, May2016_df, Jun2016_df], axis=0)
Q2_2016
```

```python
Q3_2016 = pd.concat([Jul2016_df, Aug2016_df, Sep2016_df], axis=0)
Q3_2016
```

```python
Q4_2016 = pd.concat([Oct2016_df, Nov2016_df, Dec2016_df], axis=0)
Q4_2016
```

```python
all_2016 = pd.concat([Q1_2016, Q2_2016, Q3_2016, Q4_2016], axis=0)
all_2016
```
